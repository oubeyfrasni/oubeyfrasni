### Visualisation des mots communs dans la colonne "Name"

On utilisons les librairies de NLP on peut supprimer et décomposer les mots facilement

```python
from collections import Counter
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

# Supprimer les stop_words dans la colonne "Name"
AB_NYC_2019.name = AB_NYC_2019.name.apply(lambda x: ' '.join([word for word in word_tokenize(str(x)) if word not in stop_words]))
```

```python
# On récupère les mots décomposées de chaque ligne et on aplatit les résultats sous un numpy.ndarray
words_tokenizer = AB_NYC_2019.name.apply(lambda x: [word for word in word_tokenize(str(x))]).ravel()
```

```python
import itertools

# On rend une liste dans les listes en une seule liste
words_tokenizer_list = list(itertools.chain(*words_tokenizer.tolist()))
```

```python
# On supprimons les mots non utiles
list_to_delete = [",", "!", "1", "2", ".", "&", "-"]

top_words = Counter(words_tokenizer_list)

for i in list_to_delete:
    del top_words[i]

# On laisse juste les 20 premières
top_words = top_words.most_common()[0:20]

# Mettre les résultats sous un DataFrame pour faciliter la visualisation
df_w = pd.DataFrame(top_words)
df_w.rename(columns={0:'Words', 1:'Count'}, inplace=True)

viz_9 = sns.barplot(x='Words', y='Count', data=df_w)
viz_9.set_title('20 noms utilisées dans liste des noms',fontsize=16)
viz_9.set_ylabel('Nombre de mots',fontsize=14)
viz_9.set_xlabel('Mots',fontsize=14)
viz_9.set_xticklabels(viz_9.get_xticklabels(), rotation=80)

plt.savefig("./plot/count_most_commun_used_names.png",dpi = 300)
```

# Preprocessing

```python
airbnb = AB_NYC_2019.copy()
# Supprimons les colonnes non utilisables
airbnb.drop(columns=["name", "host_id", "host_name","last_review"],inplace=True)

# Replaçons les valeurs manquantes par des zéros
airbnb.reviews_per_month = airbnb.reviews_per_month.replace(np.nan,0)
airbnb.number_of_reviews = airbnb.number_of_reviews.replace(np.nan,0)
```

```python
# Encodage des colonnes neighbourhood, neighbourhood_group et room_type; voir le prochaine notebook pour autre méthodes
airbnb.neighbourhood_group = airbnb.neighbourhood_group.factorize()[0]
airbnb.neighbourhood = airbnb.neighbourhood.factorize()[0]
airbnb.room_type = airbnb.room_type.factorize()[0]

# Tri les données par prix
airbnb.sort_values(by='price',ascending=True,inplace=True)

# Supprission des valeurs manquantes
airbnb.dropna(inplace=True)
# Sauvegarder les données préprocessés
#airbnb.to_csv('./data/processed/preprocessing_high_data.csv', index=False)
```

# **Modélisation**

```python
from sklearn.model_selection import train_test_split
```

## Test/Train split

```python
# Séparation de 'price_log' comme la colonne qu'on veut la prédire et les autres colonnes
response = 'price_log'
features = [col for col in airbnb if col != response and col != 'price']

# On prend 30% des données testés
X_train, X_test, y_train, y_test = train_test_split(
    airbnb[features],
    airbnb[response],
    test_size=0.3
)
```

## Régression Linéaire

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(copy_X= True, fit_intercept = True, normalize = True)
model.fit(X_train, y_train)
```

```python
predicts = model.predict(X_test)
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

# Implémantation du RMSE et R² scoring on ajoutons l'exponentiel sur y_test et predicts
print("Root mean squared error est:")
print(round(np.sqrt(metrics.mean_squared_error(np.exp(y_test),np.exp(predicts))),2))

print('R2 score est:')
r2 = r2_score(np.exp(y_test),np.exp(predicts))
print(round(r2,2))
```

```python
# Affichage des valeurs actuels avect les prédits dans les données de test 
error_airbnb = pd.DataFrame({
        'Valeurs actuels': np.array(np.exp(y_test)).flatten(),
        'Valeurs prédits': np.round(np.exp(predicts).flatten(),2)})

error_airbnb
```

```python
# visualisation prédiction vs actual de log_price
sns.regplot(predicts,y_test,marker="+",ci=100)
plt.xlabel('Prédictions',fontsize=12, weight="bold")
plt.ylabel('Actual',fontsize=12, weight="bold")
plt.title("Prédiction avec linéare régression",fontsize=15, weight="bold")
plt.show()
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
# A chaque itération on enregistre le progrès des scorings sur le test et le train data 
progress = dict({
    'train':{
        'rmse':[],
        'r2_score':[]
    },
    'test':{
        'rmse':[],
        'r2_score':[]
    },
})

# Itération de 100 profondeurs
for i in range(1,100):
    dtree = DecisionTreeRegressor(max_depth=i)

    dtree.fit(X_train, y_train)
    pred_train_tree= dtree.predict(X_train)
    progress['train']['rmse'].append(np.sqrt(mean_squared_error(np.exp(y_train),np.exp(pred_train_tree))))
    progress['train']['r2_score'].append(r2_score(np.exp(y_train), np.exp(pred_train_tree)))
    
    pred_test_tree= dtree.predict(X_test)
    progress['test']['rmse'].append(np.sqrt(mean_squared_error(np.exp(y_test),np.exp(pred_test_tree))))
    progress['test']['r2_score'].append(r2_score(np.exp(y_test), np.exp(pred_test_tree)))
```

```python
mlviz_1 = sns.lineplot(range(1,100),progress['test']['rmse'])
sns.lineplot(range(1,100),progress['train']['rmse'], ax = mlviz_1)
mlviz_1.set_xlabel("number of depth", fontsize = 14)
mlviz_1.set_ylabel("RMSE", fontsize = 14)
plt.legend(labels=['Test', 'Train'],fontsize=14)
```

```python
mlviz_2 = sns.lineplot(range(1,100),progress['test']['r2_score'])
sns.lineplot(range(1,100),progress['train']['r2_score'], ax = mlviz_2)
mlviz_2.set_xlabel("number of depth", fontsize = 14)
mlviz_2.set_ylabel("R2 score", fontsize = 14)
plt.legend(labels=['Test', 'Train'],fontsize=14)
```

```python
error_airbnb = pd.DataFrame({
        'Actual Values': np.array(np.exp(y_test)).flatten(),
        'Predicted Values': np.exp(pred_test_tree).flatten()})

error_airbnb
```
